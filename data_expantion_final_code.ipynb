```python
import pandas as pd
import numpy as np
from tqdm import tqdm
import os

# Function to generate file paths for a given prefix (hc_1, hc_5, or hc_10)
def generate_file_paths(prefix):
    base_files = [
        '2000.csv', '2100.csv', '2200.csv',
        '2300.csv', '2400.csv', '2500.csv',
        '2600.csv', '2700.csv', '2800.csv',
        '2900.csv', '3000.csv'
    ]
    return [f'data/filtered_{prefix}_{f}' for f in base_files]

# Generate file paths for hc_1, hc_5, and hc_10
file_paths_hc1 = generate_file_paths('hc_1')
file_paths_hc5 = generate_file_paths('hc_5')
file_paths_hc10 = generate_file_paths('hc_10')

# Combine all file paths
all_file_paths = file_paths_hc1 + file_paths_hc5 + file_paths_hc10

# List of PID filters
pid_filters = {
    'PID 11 and 13': [11, 13],
    'PID -11 and -13': [-11, -13],
    'PID 5': [5]
}

def extract_correlation_pairs(correlation_matrix, pairs):
    """Extract specified pairs from the upper diagonal of the correlation matrix."""
    pair_values = {}
    for pair in pairs:
        feature1, feature2 = pair
        if feature1 in correlation_matrix.columns and feature2 in correlation_matrix.columns:
            pair_values[pair] = correlation_matrix.loc[feature1, feature2]
    return pair_values

# Define the pairs to extract
pairs_to_extract = [
    ('PT', 'Pz'),
    ('PT', 'Eta'),
    ('PT', 'Phi'),
    ('Pz', 'Eta'),
    ('Pz', 'Phi'),
    ('Eta', 'Phi')
]

# Process each file
for file_path in tqdm(all_file_paths, desc="Processing datasets"):
    if not os.path.exists(file_path):
        print(f"File not found: {file_path}, skipping...")
        continue

    # Read the filtered CSV file
    filtered_df = pd.read_csv(file_path)

    # Process each PID filter
    for pid_label, pid_values in pid_filters.items():
        # Filter the dataframe for the current PID values
        filtered_signal = filtered_df[filtered_df['PID'].isin(pid_values)]

        # Define the features to compute correlation for
        features = ['PT', 'Pz', 'Eta', 'Phi']

        # Compute the correlation matrix
        correlation_matrix = filtered_signal[features].corr()

        # Extract the correlation values for the specified pairs
        pair_values = extract_correlation_pairs(correlation_matrix, pairs_to_extract)

        # Get the dataset label from the file path
        dataset_label = file_path.split('/')[-1].replace('.csv', '')

        # Print the results
        print(f"\nCorrelation values for {dataset_label} with {pid_label}:")
        for pair, value in pair_values.items():
            print(f"{pair}: {value}")
```

```python
import pandas as pd
import numpy as np
from tqdm import tqdm

# List of CSV file paths for all batches
batches = ['hc_1', 'hc_5', 'hc_10']
file_paths = {
    'hc_1': [
        'data/filtered_hc_1_2000.csv',
        'data/filtered_hc_1_2100.csv',
        'data/filtered_hc_1_2200.csv',
        'data/filtered_hc_1_2300.csv',
        'data/filtered_hc_1_2400.csv',
        'data/filtered_hc_1_2500.csv',
        'data/filtered_hc_1_2600.csv',
        'data/filtered_hc_1_2700.csv',
        'data/filtered_hc_1_2800.csv',
        'data/filtered_hc_1_2900.csv',
        'data/filtered_hc_1_3000.csv',
    ],
    'hc_5': [
        'data/filtered_hc_5_2000.csv',
        'data/filtered_hc_5_2100.csv',
        'data/filtered_hc_5_2200.csv',
        'data/filtered_hc_5_2300.csv',
        'data/filtered_hc_5_2400.csv',
        'data/filtered_hc_5_2500.csv',
        'data/filtered_hc_5_2600.csv',
        'data/filtered_hc_5_2700.csv',
        'data/filtered_hc_5_2800.csv',
        'data/filtered_hc_5_2900.csv',
        'data/filtered_hc_5_3000.csv',
    ],
    'hc_10': [
        'data/filtered_hc_10_2000.csv',
        'data/filtered_hc_10_2100.csv',
        'data/filtered_hc_10_2200.csv',
        'data/filtered_hc_10_2300.csv',
        'data/filtered_hc_10_2400.csv',
        'data/filtered_hc_10_2500.csv',
        'data/filtered_hc_10_2600.csv',
        'data/filtered_hc_10_2700.csv',
        'data/filtered_hc_10_2800.csv',
        'data/filtered_hc_10_2900.csv',
        'data/filtered_hc_10_3000.csv',
    ]
}

# List of PID filters with corresponding output file names
pid_filters = {
    'PID 11 and 13': {'pids': [11, 13], 'file_name': 'correlation_results_pid_11_13.csv'},
    'PID -11 and -13': {'pids': [-11, -13], 'file_name': 'correlation_results_pid_-11_-13.csv'},
    'PID 5': {'pids': [5], 'file_name': 'correlation_results_pid_5.csv'}
}

def extract_correlation_pairs(correlation_matrix, pairs):
    """Extract specified pairs from the upper diagonal of the correlation matrix."""
    pair_values = {}
    for pair in pairs:
        feature1, feature2 = pair
        if feature1 in correlation_matrix.columns and feature2 in correlation_matrix.columns:
            pair_values[pair] = correlation_matrix.loc[feature1, feature2]
    return pair_values

# Define the pairs to extract
pairs_to_extract = [
    ('PT', 'Pz'),
    ('PT', 'Eta'),
    ('PT', 'Phi'),
    ('Pz', 'Eta'),
    ('Pz', 'Phi'),
    ('Eta', 'Phi')
]

# Initialize a dictionary to store output dataframes for each PID filter
output_dfs = {
    'PID 11 and 13': pd.DataFrame(columns=['batch', 'mass', 'R_Pt_Pz', 'R_Pt_Eta', 'R_Pt_Phi', 'R_Pz_Eta', 'R_Pz_Phi', 'R_Eta_Phi']),
    'PID -11 and -13': pd.DataFrame(columns=['batch', 'mass', 'R_Pt_Pz', 'R_Pt_Eta', 'R_Pt_Phi', 'R_Pz_Eta', 'R_Pz_Phi', 'R_Eta_Phi']),
    'PID 5': pd.DataFrame(columns=['batch', 'mass', 'R_Pt_Pz', 'R_Pt_Eta', 'R_Pt_Phi', 'R_Pz_Eta', 'R_Pz_Phi', 'R_Eta_Phi'])
}

# Process each batch and file
for batch, paths in file_paths.items():
    for i, file_path in enumerate(tqdm(paths, desc=f"Processing datasets for {batch}")):
        # Determine the mass value based on the iteration
        mass_value = 2000 + 100 * i

        # Read the filtered CSV file
        filtered_df = pd.read_csv(file_path)

        # Process each PID filter
        for pid_label, pid_info in pid_filters.items():
            # Filter the dataframe for the current PID values
            filtered_signal = filtered_df[filtered_df['PID'].isin(pid_info['pids'])]

            # Define the features to compute correlation for
            features = ['PT', 'Pz', 'Eta', 'Phi']

            # Compute the correlation matrix
            correlation_matrix = filtered_signal[features].corr()

            # Extract the correlation values for the specified pairs
            pair_values = extract_correlation_pairs(correlation_matrix, pairs_to_extract)

            # Create a dictionary for the new row
            row = {
                'batch': batch,
                'mass': mass_value,
                'R_Pt_Pz': pair_values.get(('PT', 'Pz')),
                'R_Pt_Eta': pair_values.get(('PT', 'Eta')),
                'R_Pt_Phi': pair_values.get(('PT', 'Phi')),
                'R_Pz_Eta': pair_values.get(('Pz', 'Eta')),
                'R_Pz_Phi': pair_values.get(('Pz', 'Phi')),
                'R_Eta_Phi': pair_values.get(('Eta', 'Phi'))
            }

            # Append the new row to the appropriate output dataframe using pd.concat
            output_dfs[pid_label] = pd.concat([output_dfs[pid_label], pd.DataFrame([row])], ignore_index=True)

# Save the output dataframes to their respective CSV files
for pid_label, df in output_dfs.items():
    df.to_csv(pid_filters[pid_label]['file_name'], index=False)
    print(f"\nCorrelation results saved to '{pid_filters[pid_label]['file_name']}'")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# File paths for the generated CSV files
csv_files = {
    'PID 11 and 13': 'correlation_results_pid_11_13.csv',
    'PID -11 and -13': 'correlation_results_pid_-11_-13.csv',
    'PID 5': 'correlation_results_pid_5.csv'
}

# Variables to plot
variables = ['R_Pt_Pz', 'R_Pt_Eta', 'R_Pt_Phi', 'R_Pz_Eta', 'R_Pz_Phi', 'R_Eta_Phi']

# Horizontal lines for specific subplots
horizontal_lines = {
    ('R_Pt_Pz', 'PID 11 and 13'): 0.02571,
    ('R_Pt_Pz', 'PID -11 and -13'): 0.014458,
    ('R_Pt_Pz', 'PID 5'): -0.012316,
    ('R_Pt_Eta', 'PID 11 and 13'): -0.00220,
    ('R_Pt_Eta', 'PID -11 and -13'): -0.002730,
    ('R_Pt_Eta', 'PID 5'): -0.000265,
    ('R_Pt_Phi', 'PID 11 and 13'): 0.00321,
    ('R_Pt_Phi', 'PID -11 and -13'): 0.003407,
    ('R_Pt_Phi', 'PID 5'): 0.012313,
    ('R_Pz_Eta', 'PID 11 and 13'): 0.40031,
    ('R_Pz_Eta', 'PID -11 and -13'): 0.402572,
    ('R_Pz_Eta', 'PID 5'): 0.1464155,
    ('R_Pz_Phi', 'PID 11 and 13'): -0.00601,
    ('R_Pz_Phi', 'PID -11 and -13'): -0.015739,
    ('R_Pz_Phi', 'PID 5'): -0.003294,
    ('R_Eta_Phi', 'PID 11 and 13'): -0.01012,
    ('R_Eta_Phi', 'PID -11 and -13'): -0.012619,
    ('R_Eta_Phi', 'PID 5'): 0.003239
}

# Number of rows and columns for subplots
n_rows = len(variables)
n_cols = len(csv_files)

# Create a single figure with multiple subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(5 * n_cols, 4 * n_rows), sharex=False, sharey=False)

# Ensure axes is a 2D array for consistent indexing
if n_rows == 1 and n_cols == 1:
    axes = [[axes]]
elif n_rows == 1:
    axes = [axes]
elif n_cols == 1:
    axes = [[ax] for ax in axes]

# Iterate over variables and PID labels to populate subplots
for row_idx, var in enumerate(variables):
    for col_idx, (pid_label, file_path) in enumerate(csv_files.items()):
        ax = axes[row_idx][col_idx]
        
        # Read the CSV file
        df = pd.read_csv(file_path)
        
        # Iterate over each batch and plot the data
        for batch in df['batch'].unique():
            # Filter data by batch
            batch_df = df[df['batch'] == batch]
            
            # Plot the line for this batch
            ax.plot(batch_df['mass'], batch_df[var], marker='o', linestyle='--', label=f'Batch {batch}')
        
        # Add the horizontal line with the same color and label 'ttbar'
        if (var, pid_label) in horizontal_lines:
            ax.axhline(y=horizontal_lines[(var, pid_label)], color='magenta', linestyle='--', linewidth=1, label='ttbar')
        
        # Set labels and title
        ax.set_xlabel('Mass')
        ax.set_ylabel(var)
        ax.set_title(f'{var} vs Mass\n({pid_label})')
        
        # Disable grid lines
        ax.grid(False)
        
        # Add legend only once for the ttbar line to avoid duplicates
        if col_idx == n_cols - 1 and (var, pid_label) in horizontal_lines:
            ax.legend()

# Adjust layout to prevent overlap
plt.tight_layout()

# Save the figure as a PNG file
plot_filename = 'all_line_plots_with_ttbar_horizontal_lines.png'
plt.savefig(plot_filename)
plt.show()

print(f"All line plots with 'ttbar' horizontal lines saved in a single figure: {plot_filename}")
```
