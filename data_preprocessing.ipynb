```python
import pandas as pd
import numpy as np
from tqdm import tqdm

# List of CSV file paths
file_paths = [
    'data/filtered_hc_1_2000.csv',
    'data/filtered_hc_1_2100.csv',
    'data/filtered_hc_1_2200.csv',
    'data/filtered_hc_1_2300.csv',
    'data/filtered_hc_1_2400.csv',
    'data/filtered_hc_1_2500.csv',
    'data/filtered_hc_1_2600.csv',
    'data/filtered_hc_1_2700.csv',
    'data/filtered_hc_1_2800.csv',
    'data/filtered_hc_1_2900.csv',
    'data/filtered_hc_1_3000.csv'
]

# List of PID filters with corresponding output file names
pid_filters = {
    'PID 11 and 13': {'pids': [11, 13], 'file_name': 'correlation_results_pid_11_13.csv'},
    'PID -11 and -13': {'pids': [-11, -13], 'file_name': 'correlation_results_pid_-11_-13.csv'},
    'PID 5': {'pids': [5], 'file_name': 'correlation_results_pid_5.csv'}
}

def extract_correlation_pairs(correlation_matrix, pairs):
    """Extract specified pairs from the upper diagonal of the correlation matrix."""
    pair_values = {}
    for pair in pairs:
        feature1, feature2 = pair
        if feature1 in correlation_matrix.columns and feature2 in correlation_matrix.columns:
            pair_values[pair] = correlation_matrix.loc[feature1, feature2]
    return pair_values

# Define the pairs to extract
pairs_to_extract = [
    ('PT', 'Pz'),
    ('PT', 'Eta'),
    ('PT', 'Phi'),
    ('Pz', 'Eta'),
    ('Pz', 'Phi'),
    ('Eta', 'Phi')
]

# Initialize a dictionary to store output dataframes for each PID filter
output_dfs = {
    'PID 11 and 13': pd.DataFrame(columns=['mass', 'R_Pt_Pz', 'R_Pt_Eta', 'R_Pt_Phi', 'R_Pz_Eta', 'R_Pz_Phi', 'R_Eta_Phi']),
    'PID -11 and -13': pd.DataFrame(columns=['mass', 'R_Pt_Pz', 'R_Pt_Eta', 'R_Pt_Phi', 'R_Pz_Eta', 'R_Pz_Phi', 'R_Eta_Phi']),
    'PID 5': pd.DataFrame(columns=['mass', 'R_Pt_Pz', 'R_Pt_Eta', 'R_Pt_Phi', 'R_Pz_Eta', 'R_Pz_Phi', 'R_Eta_Phi'])
}

# Process each file
for i, file_path in enumerate(tqdm(file_paths, desc="Processing datasets")):
    # Determine the mass value based on the iteration
    mass_value = 2000 + 100 * i

    # Read the filtered CSV file
    filtered_df = pd.read_csv(file_path)

    # Process each PID filter
    for pid_label, pid_info in pid_filters.items():
        # Filter the dataframe for the current PID values
        filtered_signal = filtered_df[filtered_df['PID'].isin(pid_info['pids'])]

        # Define the features to compute correlation for
        features = ['PT', 'Pz', 'Eta', 'Phi']

        # Compute the correlation matrix
        correlation_matrix = filtered_signal[features].corr()

        # Extract the correlation values for the specified pairs
        pair_values = extract_correlation_pairs(correlation_matrix, pairs_to_extract)

        # Create a dictionary for the new row
        row = {
            'mass': mass_value,
            'R_Pt_Pz': pair_values.get(('PT', 'Pz')),
            'R_Pt_Eta': pair_values.get(('PT', 'Eta')),
            'R_Pt_Phi': pair_values.get(('PT', 'Phi')),
            'R_Pz_Eta': pair_values.get(('Pz', 'Eta')),
            'R_Pz_Phi': pair_values.get(('Pz', 'Phi')),
            'R_Eta_Phi': pair_values.get(('Eta', 'Phi'))
        }

        # Append the new row to the appropriate output dataframe using pd.concat
        output_dfs[pid_label] = pd.concat([output_dfs[pid_label], pd.DataFrame([row])], ignore_index=True)

        # Print the results
        print(f"\nCorrelation values for {file_path.split('/')[-1].replace('.csv', '')} with {pid_label}:")
        for pair, value in pair_values.items():
            print(f"{pair}: {value}")

# Save the output dataframes to their respective CSV files
for pid_label, df in output_dfs.items():
    df.to_csv(pid_filters[pid_label]['file_name'], index=False)
    print(f"\nCorrelation results saved to '{pid_filters[pid_label]['file_name']}'")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C
import numpy as np

# File paths for the generated CSV files
csv_files = {
    'PID 11 and 13': 'correlation_results_pid_11_13.csv',
    'PID -11 and -13': 'correlation_results_pid_-11_-13.csv',
    'PID 5': 'correlation_results_pid_5.csv'
}

# Variables to plot
variables = ['R_Pt_Pz', 'R_Pt_Eta', 'R_Pt_Phi', 'R_Pz_Eta', 'R_Pz_Phi', 'R_Eta_Phi']

# Iterate over each CSV file and generate plots
for pid_label, file_path in csv_files.items():
    # Read the CSV file
    df = pd.read_csv(file_path)

    # Generate plots for each variable
    for var in variables:
        plt.figure(figsize=(14, 6))

        # Create subplots
        plt.subplot(1, 2, 1)  # First subplot for original data
        plt.plot(df['mass'], df[var], 'o--', label=f'{var} data ({pid_label})')
        plt.xlabel('Mass')
        plt.ylabel(var)
        plt.title(f'{var} vs Mass ({pid_label})')
        plt.grid(True)
        plt.legend()

        # Extract data
        X = df[['mass']].values
        y = df[var].values

        # Define the kernel and GP model
        kernel = C(1.0, (1e-3, 1e3)) * RBF(1.0, (1e-2, 1e2))
        gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10)

        # Fit to data
        gp.fit(X, y)

        # Make predictions
        X_fit = np.linspace(X.min(), X.max(), 1000)[:, np.newaxis]
        y_pred, sigma = gp.predict(X_fit, return_std=True)

        # Create the second subplot for GP regression
        plt.subplot(1, 2, 2)
        plt.plot(df['mass'], df[var], 'o', label=f'{var} data ({pid_label})')
        plt.plot(X_fit, y_pred, 'r-', label='GP Prediction')
        plt.fill_between(X_fit.flatten(), y_pred - 1.96 * sigma, y_pred + 1.96 * sigma, alpha=0.2, color='r', label='95% Confidence Interval')
        plt.xlabel('Mass')
        plt.ylabel(var)
        plt.title(f'GP Regression ({var} vs Mass, {pid_label})')
        plt.grid(True)
        plt.legend()

        plt.tight_layout()

        # Save the plot as a PNG file
        plot_filename = f'{var}_vs_mass_{pid_label.replace(" ", "_")}_gp_subplots.png'
        plt.savefig(plot_filename)
        plt.show()

        print(f"Plot saved: {plot_filename}")
```
